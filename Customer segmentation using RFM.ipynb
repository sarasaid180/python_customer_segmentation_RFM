#  RFM Analysis

Here, you are going to perform following opertaions:
 * For Recency, Calculate the number of days between present date and date of last purchase each customer.
 * For Frequency, Calculate the number of orders for each customer.
 * For Monetary, Calculate sum of purchase price for each customer.
 
## Creating RFM Dataframe

```python
 # RFM 
print( df ['InvoiceDate'].min(), df['InvoiceDate'].max())  # let's see the latest day 
```

```python
now =  dt.datetime(2011,12,10) 
now 
```

###  Groupby and aggregation

The .agg() method can be used with a tuple or list of aggregations as input. When applying multiple aggregations on multiple columns, the aggregated DataFrame has a multi-level column index.

```python
 # RFM Dataframe 
# lambda function is a small anonymous function 
# lambda function can take any number of arguments, but can only have one expression. 
# Syntax==> lambda arguments : expression (expression is executed and the result is returned) 
rfm = df.groupby('CustomerID').agg({'InvoiceDate' : lambda day : (now - day.max()).days, 
                                    'InvoiceNo': lambda num : len(num), 
                                    'Total_Price': lambda price : price.sum()}) 
 
 
rfm
```

```python
 # Change Cloumns names 
col_list = ['Recency','Frequency','Monetary'] 
 
rfm.columns = col_list 
rfm 
```

```python
# Computing Quartiles of RFM values. 
# qcut () is Quartile -based discretization function.
# ex:1000 values for values for 5 quartiles 
# Note that in the following code, we reversed the numbers for Recency because lower Recency is better for us 
 
rfm["R"] = pd.qcut(rfm["Recency"],5,labels=[5,4,3,2,1]) 
rfm["F"] = pd.qcut(rfm["Frequency"],5,labels=[1,2,3,4,5]) 
rfm["M"] = pd.qcut(rfm["Monetary"],5,labels=[1,2,3,4,5]) 
# Calculating RFM Scores 
rfm["RFM_Score"] = rfm["R"].astype(str) +rfm["F"].astype(str) + rfm["M"].astype(str) 
rfm.head() 
```

```python
# Customer Segmentation by Recency and Frequency 
# Our dataset is ready for segmenting our customers. 
seg_map = { 
    r'[1-2][1-2]': 'Hibernating',
    r'[1-2][3-4]': 'At Risk', 
    r'[1-2]5': 'Can\'t Loose', 
    r'3[1-2]': 'About to Sleep', 
    r'33': 'Need Attention', 
    r'[3-4][4-5]': 'Loyal Customers', 
    r'41': 'Promising', 
    r'51': 'New Customers', 
    r'[4-5][2-3]': 'Potential Loyalists', 
    r'5[4-5]': 'Champions' 
} 
```

```python
 # We determined how to name our segments using regex and now we are ready to create our segment column. 
 
rfm['Segment'] = rfm['R'].astype(str) + rfm['F'].astype(str) 
 
rfm['Segment'] = rfm['Segment'].replace(seg_map, regex=True) 
rfm.head() 
```

```python
 # average values of each segment 
rfm.groupby('Segment').mean().sort_values('Monetary')
```
